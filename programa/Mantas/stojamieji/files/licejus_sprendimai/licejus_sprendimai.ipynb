## Sprendimo ir vertinimo pavyzdys (2007m.)
1\. Petro sieninis laikrodis per valandą nuskuba 2 minutes, o žadintuvas per valandą vėluoja 1 minutę.
Vakare Petras tiksliai nustatė abu laikrodžius. Rytą prabudus, sieninis laikrodis rodė 7 valandas 30
minučių, o žadintuvas – 7 valandas. Kiek laiko buvo iš tikrųjų? (4tšk.)

**Sprendimas.**
  * Laikrodžių parodymų skirtumas ryte: 7h 30min - 7h = 30(min.)
  * Sudarome proporciją žadintuvo atsilikimo laikui nustatyti: $\boxed{\begin{array}{rcl} 30 & - & 3 \\ x & - & 1 \end{array}}$
  * Proporcijos sprendimas: $30=3x \Rightarrow x=10$
  * Kiek laiko buvo iš tikrųjų: $7\text{h} + 10\text{min} = 7\text{h }10\text{min}$
   
**Vertinimas.** Kiekviena daliai po tašką. Proporcijos sudarymą ir sprendimą galima pakeisti kitais žingsniais:
  * Kiek valandų praėjo nuo nustatymo: 30 / 3 = 10(h)
  * Kiek per jas atsiliko žadintuvas: 1 $\times$ 10 = 10(min.)

2\. Viena sekretorė gali perrašyti rankraštį per 5 h 20 min., o kita per 4 h 40 min. Dirbdamos kartu jos
perrašė 90 puslapių. Kiek puslapių perrašė kiekviena sekretorė? (3tšk.)

**Sprendimas.**
  * Sekretorių darbo tempų santykis: 5h 20min : 4h 40min. = 320min : 280min = 8 : 7
  * Sudarome lygtį, kurioje $7x$ ir $8x$ atitinka sekretorių darbo tempus: $\boxed{7x + 8x = 90}$
  * Išsprendžiame lygtį, randame atsakymus: $7x=42$(psl.), $8x=48$(psl.)
  
**Vertinimas.** Kiekviena daliai po tašką. Lygties sudarymą ir sprendimą galima pakeisti kitais žingsniais:
  * Kurią dalį darbo atliko kiekviena sekretorė: $\frac{7}{15}$ ir $\frac{8}{15}$
  * Apskaičiuoti tas dalis: $90 \times \frac{7}{15} = 42$ ir $90 \times \frac{8}{15} = 48$

3\. Apskaičiuokite: $\displaystyle \sqrt{\frac{8^{10}+4^{10}}{8^4+4^{11}}}$ (3tšk.)

**Sprendimas.** $\displaystyle \sqrt{\frac{8^{10}+4^{10}}{8^4+4^{11}}} = \sqrt{\frac{\left(2^3\right)^{10}+\left(2^2\right)^{10}}{\left(2^3\right)^4+\left(2^2\right)^{11}}} = \boxed{\sqrt{\frac{2^{30}+2^{20}}{2^{12}+2^{22}}}} = \boxed{\sqrt{\frac{2^{20}\left(1+2^{10}\right)}{2^{12}\left(1+2^{10}\right)}}} = \sqrt{\frac{2^{20}}{2^{12}}}=\sqrt{2^8}=2^4=\boxed{16}$

**Vertinimas.** Po tašką už kiekvieną apibrauktą dalį.

4\. Apskaičiuokite reiškinio $\frac{n^2-mn}{4m^2}$ skaitinę reikšmę, jeigu žinoma, kad $\frac{n-2m}{m}=3$ (3tšk.)

**Sprendimas.**
* $\boxed{\frac{n-2m}{m}=3} \Rightarrow \boxed{\frac{n}{m}-\frac{2m}{m}=3} \Rightarrow \boxed{\frac{n}{m}-2=3} \Rightarrow  \boxed{\frac{n}{m}=5} \Rightarrow \boxed{n=5m}$
* $\displaystyle \frac{n^2-mn}{4m^2} = \frac{(5m)^2-m\cdot 5m}{4m^2} = \frac{25m^2-5m^2}{4m^2} = \frac{20m^2}{4m^2} = 5$

**Vertinimas.** 1tšk. už radimą, kaip priklauso $n$ nuo $m$. 1tšk. už šios priklausomybės panaudojimą skaičiavimuose. 1tšk. už teisingus skaičiavimus.

5\. Skaičius $a$ didesnis $20\%$ už skaičių $b$. Keliais procentais skaičius $b$ mažesnis už skaičių $a$? (3tšk.)

**Sprendimas.**
* $a$ sudaro $120\%$ skaičiaus $b$ (ir 100% savęs).
* Dabar tegu $100\%$ atitinka skaičių $b$. Kiek $\%$ atitiks skaičių $a$? Galima sudaryti proporciją: $\boxed{\begin{array}{rcl} 100\% & - & 120\% \\ x\% & - & 100\% \end{array}}$
* Teisingai išsprendę proporciją gautume $83\frac{1}{3}\%$. Vadinasi $b$ yra $16\frac{2}{3}\%$ mažesnis už $a$.

**Vertinimas.** Po tašką už kiekvieną dalį.

6\. Trikampio $ABC$ kampai $\angle A = 55^o$, $\angle C = 75^o$. Kraštinėse $AB$ ir $BC$ atitinkamai pažymėti taškai $D$ ir
$E$ taip, kad $DB = BE$ . Raskite kampo $\angle ADE$ didumą. (4tšk.)

![](imgeo2007.PNG) | **DUOTA:** <ul><li> $\angle A = 55^o$ </li><li>$\angle C = 75^o$ </li><li> $DB = BE$ </ul> **RASTI:** <ul><li>$\angle ADE$</li></ul> 
:---|:---

**Sprendimas.** 
* $\angle B = 180^o - \angle A - \angle C = 180^o - 55^o - 75^o = 50^o$
* $\boxed{\angle BDE + \angle BED} = 180^o - \angle B = 180^o - 50^o = 130^o$
* $\triangle{DBE}$ lygiašonis, todėl $\displaystyle\angle BDE = \angle BED = \frac{\boxed{\angle BDE + \angle BED}}{2} = \frac{130^o}{2}=65^o$
* $\angle ADE = 180^o - \angle{BED} = 180^o - 65^o = 115^o$
