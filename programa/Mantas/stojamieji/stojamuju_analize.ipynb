  * Laikrodžių parodymų skirtumas per valandą: 2 + 1 = 3(min.)
  * Laikrodžių parodymų skirtumas ryte: 7h 30min - 7h = 30(min.)

  * Kiek per jas atsiliko žadintuvas: 1 $\times$ 10 = 10(min.)
  * Kiek laiko buvo iš tikrųjų: 7h + 10min = 7h 10 (min.)

2\. Viena sekretorė gali perrašyti rankraštį per 5 h 20 min., o kita per 4 h 40 min. Dirbdamos kartu jos
perrašė 90 puslapių. Kiek puslapių perrašė kiekviena sekretorė? (3tšk.)

**Sprendimas.**
  * Sekretorių darbo tempų santykis: 5h 20min : 4h 40min. = 320min : 280min = 8 : 7
  * Sudarome lygtį, kurioje $7x$ ir $8x$ atitinka sekretorių darbo tempus: $\boxed{7x + 8x = 90}$
  * Išsprendžiame lygtį, randame atsakymus: $7x=42$(psl.), $8x=48$(psl.)
  
**Pastabos.** Galima nesudaryti lygties. Tada reiktų teisingai nustatyti, kurią dalį darbo atliko kiekviena sekretorė: $\frac{7}{15}$ ir $\frac{8}{15}$. Po to apskaičiuoti tas dalis: $90 \times \frac{7}{15} = 42$ ir $90 \times \frac{8}{15} = 48$

3\. Apskaičiuokite: $\sqrt{\frac{8^{10}+4^{10}}{8^4+4^{11}}}$ (3tšk.)

**Sprendimas.** $\displaystyle \sqrt{\frac{8^{10}+4^{10}}{8^4+4^{11}}} = \sqrt{\frac{\left(2^3\right)^{10}+\left(2^2\right)^{10}}{\left(2^3\right)^4+\left(2^2\right)^{11}}} = \boxed{\sqrt{\frac{2^{30}+2^{20}}{2^{12}+2^{22}}}} = \boxed{\sqrt{\frac{2^{20}\left(1+2^{10}\right)}{2^{12}\left(1+2^{10}\right)}}} = \sqrt{\frac{2^{20}}{2^{12}}}=\sqrt{2^8}=2^4=\boxed{16}$

**Pastabos.** Apibrauktos dalys rodo, už ką siūlyčiau duoti po tašką.

4\. Apskaičiuokite reiškinio $\frac{n^2-mn}{4m^2}$ skaitinę reikšmę, jeigu žinoma, kad $\frac{n-2m}{m}=3$ (3tšk.)

* $\boxed{\frac{n-2m}{m}=3} \Rightarrow \boxed{\frac{n}{m}-\frac{2m}{m}=3} \Rightarrow \boxed{\frac{n}{m}-2=3} \Rightarrow  \boxed{\frac{n}{m}=5} \Rightarrow \boxed{n=5m}$
* $\displaystyle \frac{n^2-mn}{4m^2} = \frac{(5m)^2-m\cdot 5m}{4m^2} = \frac{25m^2-5m^2}{4m^2} = \frac{20m^2}{4m^2} = 5$

### Darbo planas
* Užrašau pilnus uždavinių sprendimus, jei įmanoma bandau pastebėti ir po kelis būdus spręsti.
* Nustatau už ką konkrečiai skiriamas kiekvienas taškas sprendimuose.
* Gebėjimų sąraše yra pažymėta 18 gebėjimų. Tam tikrų metų stojamajame kiekvienas taškas yra apibūdinimas tam tikru gebėjimu. Surandu, kokį gebėjimą kiekvienas taškas vertina.
* Sudarau diagramą, kuri rodo vertinamų gebėjimų pasiskirstymą. Pvz.
  * ***dalumo suvokimas*** = 0.08, 
  * ***kombinatorikos supratimas*** = 0.1
* Kalbėdamas su Mantu per daugybę užsiėmimų įvertinu ir suvedu kiekvieno gebėjimo stiprumą. Pvz. 
  * ***dalumo suvokimas*** = 0.7, 
  * ***kombinatorikos supratimas*** = 0.3
* Nustatau gebėjimų trūkumo kainą. Pvz.:
  * ***procento sąvokos suvokimo trūkumo kaina*** = $0.07 \times (1 - 0.4) = 0.042$
  * ***kombinatorikos supratimo trūkumo kaina*** = $0.1 \times (1 - 0.3) = 0.07$
* Vienas iš strategijų yra mokytis tuos gebėjimus, kurių trūkumo kaina didžiausia. Bet nebūtinai geriausia, nes trūkumą pašalinti gali užimti ne vienodai laiko. Pvz.:
  * ***jei procentus mokytis reikia 2 pamokas, tai mokymosi efektyvumas*** = $0.03/2 = 0.015$ stojamojo balo per pamoką.
  * ***jei kombinatoriką mokytis reikia 3 pamokas, tai mokymosi efektyvumas*** = $0.042/3 = 0.014$ stojamojo balo per pamoką.
  * Išvada: kombinatorikos mokytis šiuo metu apsimoka panašiai, kaip procentų.
