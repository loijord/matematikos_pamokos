## Pirma pamoka

Pirma pamoka prasidėjo nuo Inos pastabų dėl tvarkos sąsiuvinyje ir gauto 7 iš laipsnių temos. 

Nusprendėme artimiausiu daug dėmesio skirti pastangoms uždavinių sprendimų užrašyti tvarkingai. Dar priminsiu, kas itin svarbu iš ankstesnių pamokų: kiekvieno nebūtinai savarankiškai pavykusio sprendimo tikslas taip pat turi būti išsigryninti sunkiąsias vietas - faktus, be kurių uždavinys neišsispręstų arba įsidėmėti, kokios buvo klaidos.

**Tvarka...**
* Netvarkingi sprendimai: rašomi tik veiksmai ir atsakymas, kartais netvarkingu raštu ar nesilaikant aiškaus išdėstymo.
* Tvarkingi sprendimai: kiekvieno veiksmo (atlikto žingsnio) prasmę reikia paaiškinti, taip pat, jei sprendžiame tekstinį uždavinį, reiktų paaiškinti, kaip vienas žingsnis logiškai išeina iš kito. Jei nesame tikri dėl skaičiavimų, juos darome juodraštyje. 

Kartais praverčia tekstą išdėstyti blokais (taip daroma moksliniuose straipsniuose). Tai padeda matyti, kas buvo keista atliekant pertvarkymus. Dėstymo blokais pavyzdys:

\begin{eqnarray*}
(x+y)(x-y)(x^2+y^2)(x^4+y^4)(x^8+y^8) & = & (x^2-y^2)(x^2+y^2)(x^4+y^4)(x^8+y^8) = \\
& = & (x^4-y^4)(x^4+y^4)(x^8+y^8) = \\
& = & (x^8 - y^8)(x^8+y^8) = \\
& = & x^{16} - y^{16}
\end{eqnarray*}

Visą likusią pamoką skyrėme nagrinėti pirmą stojamąjį į Biržišką tvarkingai užrašant sprendimus sąsiuvinyje. Apžvelgsiu tik dalykus, prie kurių per aptarimą praleidome daugiau laiko.

1\. **Neišspręstas uždavinys per kontrolinį su įbrėžtu į kvadratą trikampiu.**
Mantas gavo uždavinį: į kvadratą įbrėžtas trikampis taip, kad jo pagrindas sutampa su kraštine. Kvadrato kraštinės ilgis lygus 4. Koks įbrėžto trikampio plotas? Tvarkingo sprendimo pavyzdys:

![](triuzd.png)

$$S_{\Delta BCE} = \frac{BC \times BA}{2} = \frac{4 \times 4 }{2} = 8$$

2\. **Kintamųjų išreiškimas kitais duotose formulėse.**

Tęsdamas kalbą apie tvarkingus sprendimus Manto paklausiau, ar jis mokėtų išreikšti kraštinę $AD$ kitais kintamaisiais (per kitus kintamuosius), kai duota, kad $S_{\Delta BCE} = \frac{CD \times AD}{2}$. Žodis ***išreikšti*** jam pasirodė neaiškus, todėl paaiškinau, kad vieno kintamojo išreiškimu kitais laikau to kintamojo užrašymą reiškiniu, kuris ***priklauso*** tik nuo kitų kintamųjų. Kitaip tariant ***išreikšti*** kitais kintamaisiais reiškia užrašyti ***priklausomybę*** nuo kitų kintamųjų. Pavyzdys: 

$$\text{Tarkime }S_{\Delta BCE} = \frac{CD \times AD}{2}$$
$$\text{Tada }\frac{S_{\Delta BCE}}{CD} = \frac{AD}{2}$$
$$\text{Tada }\frac{2\times S_{\Delta BCE}}{CD} = AD$$
$$\text{Vadinasi, }AD = \frac{2\times S_{\Delta BCE}}{CD}$$

Kitas pavyzdys, kurį apžvelgėme - pagreičio formulė. Ar aišku, kaip iš formulės $a = \frac{v - v_0}{t}$ išreikšti likusius kintamuosius? Turėjom gauti tokias išraiškas:

$$a = \frac{v - v_0}{t} \text{(duota)}$$
$$t = \frac{v - v_0}{a}$$
$$v = at + v_0$$
$$v_0 = v - at$$

* Uždavinys su šaknų prastinimu.

Mantas spręsdamas uždavinį su šaknimis bandė užrašyti: $\sqrt{18} = \sqrt{3} + \sqrt{6}$. Tai yra labai blogai, nes buvo stengtasi atspėti taisyklę ir tai nepavyko. Iš tiesų $\sqrt{18}$ pertvarkymų Mantas per pamokas dar nepradėjo (bet pradeda), tačiau praeitą pamoką nagrinėjome $\sqrt{54}$ skaidymą panaudojant žėtonus. Mantas aiškinimo neatsiminė. Uždavinio, kurį sprendėme iš stojamųjų į Biržišką, tvarkingas sprendimas:

\begin{eqnarray*}
\sqrt{18} - \sqrt{8} + \sqrt{50} + \sqrt{27} & = & \sqrt{2\cdot 3 \cdot 3} - \sqrt{2\cdot 2 \cdot 2} + \sqrt{2\cdot 5 \cdot 5} + \sqrt{3 \cdot 3 \cdot 3} = \\
& = & 3\sqrt{2} - 2\sqrt{2} + 5\sqrt{2} + 3\sqrt{3} = \\
& = & 6\sqrt{2} + 3\sqrt{3}
\end{eqnarray*}

Mantui iškilę sunkumai:

* Neprisiminė kaip skaičiuojamas bet kurio trikampio plotas, kai žinoma jo kraštinė ir į ją išvesta aukštinė. Paminėjo, kad žino kaip tai atlikti tik su stačiaisiais trikampiais. Reikėjo pakartoti užmirštą kursą:

**Trikampio ploto formulė visais atvejais yra: $S=\frac{ah}{2}$ Čia $a$ - pagrindo, į kurį išvesta aukštinė, ilgis, o $h$ - tos aukštinės ilgis:**

* Nežinau, ar suprato žodžius ***išreikšti*** ir ***priklausomybė***.
* Per kontrolinį įvėlė žioplą klaidą ($2x = x^2$)
* Taikė daug neteisingų savybių: $(a^2-b^2)^2 = a^4 - b^4$, $(a^2+b^2)^2 = a^4 + b^4$, $\sqrt{ab} = \sqrt{a} + \sqrt{b}$ **Spėlioti taisykles - labai blogai!**

Pasikartojome, kas svarbiausius faktus apie Pitagoro teoremą:

* Teorema. Jei *a*, *b* yra stačiojo trikampio statiniai, o *c* - įžambinė, tai $a^2+b^2=c^2$
* Pitagoro trejetais laikomi stačiojo trikampio kraštinių ilgiai, kurie yra sveikieji. 
* Dažniausiais pasitaikantys Pitagoro trejetai: $(3,4,5)$, $(5, 12 ,13)$.
