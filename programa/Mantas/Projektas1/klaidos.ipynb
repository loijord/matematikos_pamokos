### Painioji lingvistika
Prieš tai mokykloje teko mokytis tam tikrus galbūt vietomis sunkiai paaiškinamus dėsnius:

* <span style="color:green"> $\boxed{\text{Sudėti galime bet kuria tvarka}}$ </span> $\Leftrightarrow$ <span style="color:green"> $\boxed{a+b=b+a}$ </span>
* <span style="color:green"> $\boxed{\text{Sudėjimo eiliškumas nesvarbu}}$ </span> $\Leftrightarrow$ <span style="color:green"> $\boxed{(a+b)+c=a+(b+c)}$ </span>
* <span style="color:green"> $\boxed{\text{Dauginti galime bet kuria tvarka}}$ </span> $\Leftrightarrow$ <span style="color:green"> $\boxed{a\times b = b \times a}$ </span>
* <span style="color:green"> $\boxed{\text{Dauginimo eiliškumas nesvarbu}}$ </span> $\Leftrightarrow$ <span style="color:green"> $\boxed{(a\times b)\times c=a\times (b\times c)}$ </span>
* <span style="color:green"> $\boxed{\text{Sumos sandauga = sandaugų suma}}$ </span> $\Leftrightarrow$ <span 
style="color:green"> $\boxed{a(b+c)=ab+ac}$ </span>
* <span style="color:green"> $\boxed{\text{Sumos dalmuo = dalmenų suma}}$ </span> $\Leftrightarrow$ <span 
style="color:green"> $\boxed{\frac{b+c}{a}=\frac{b}{a}+\frac{c}{a}}$ </span>
* <span style="color:green"> $\boxed{\text{Skirtumo sandauga = sandaugų skirtumas}}$ </span> $\Leftrightarrow$ <span style="color:green"> $\boxed{a(b-c)=ab-ac}$ </span>
* <span style="color:green"> $\boxed{\text{Skirtumo dalmuo = dalmenų skirtumas}}$ </span> $\Leftrightarrow$ <span style="color:green"> $\boxed{\frac{b-c}{a}=\frac{b}{a}-\frac{c}{a}}$ </span>
* <span style="color:green"> $\boxed{\text{Sandauga, sudaryta iš trupmenų = trupmena, sudaryta iš sandaugų}}$ </span> $\Leftrightarrow$ <span style="color:green"> $\boxed{\frac{a}{b}\times \frac{c}{d} = \frac{ac}{bd}}$ </span>
* <span style="color:green"> $\boxed{\text{Sandaugos kvadratas = kvadratų sandauga}}$ </span> $\Leftrightarrow$ <span style="color:green"> $\boxed{(ab)^2=a^2b^2}$ </span>
* <span style="color:green"> $\boxed{\text{Sandaugos šaknis = šaknų sandauga}}$ </span> $\Leftrightarrow$ <span style="color:green"> $\boxed{\sqrt{ab}=\sqrt{a}\sqrt{b}}$ </span>

Nesant pakankamai gerų paaiškinimų, kas yra ir kur pasitaiko sumos, sandaugos, trupmenos, kvadratai, šaknys ir kitos sąvokos bei kokios situacijos atspindi šias taisykles, vis labiau norisi jas atsiminti kaip simbolių pozicijų apkeitimus, kur dažnai galioja lingvistiniai operacijų apgręžimo dėsniai, panašūs į prieš tai paminėtus. Bandant tokiu būdu atsiminti matematines taisykles, išimčių sąrašas ima pranokti taisyklių sąrašą:

* <span style="color:red"> $\boxed{\text{Atimti bet kuria tvarka negalime!}}$ </span> $\Leftrightarrow$ <span style="color:red"> $\boxed{a-b\neq b-a}$ </span>, bet <span style="color:green"> $\boxed{a-b=-(b-a)}$ </span>
* <span style="color:red"> $\boxed{\text{Atėmimo eiliškumas svarbu!}}$ </span> $\Leftrightarrow$ <span style="color:red"> $\boxed{a-(b-c)\neq a-b-c}$ </span>, bet <span style="color:green"> $\boxed{a-(b-c) = a-b+c}$ </span>
* <span style="color:red"> $\boxed{\text{Dalinti bet kuria tvarka negalime!}}$ </span> $\Leftrightarrow$ <span style="color:red"> $\boxed{\frac{a}{b}\neq \frac{b}{a}}$ </span>, bet <span style="color:green"> $\boxed{\frac{a}{b}= \left(\frac{b}{a}\right)^{-1}}$ </span>
* <span style="color:red"> $\boxed{\text{Dalinimo eiliškumas svarbu!}}$ </span> $\Leftrightarrow$ <span style="color:red"> $\boxed{a:(b:c)\neq a:b:c}$ </span>, bet <span style="color:green"> $\boxed{a:(b:c) = a:b\times c}$ </span>
* <span style="color:red"> $\boxed{\text{Skaičiaus ir sumos dalmuo} \neq \text{skaičiaus dalmenų suma}}$ </span> $\Leftrightarrow$ <span style="color:red"> $\boxed{\frac{a}{b+c} \neq \frac{a}{b}+\frac{a}{c}}$ </span> (tolimesni pertvarkymai negalimi)
* <span style="color:red"> $\boxed{\text{Skaičiaus ir skirtumo dalmuo} \neq \text{skaičiaus dalmenų skirtumas}}$ </span> $\Leftrightarrow$ <span style="color:red"> $\boxed{\frac{a}{b-c} \neq \frac{a}{b}-\frac{a}{c}}$ </span> (tolimesni pertvarkymai negalimi)
* <span style="color:red"> $\boxed{\text{Suma, sudaryta iš trupmenų}\neq\text{trupmena, sudaryta iš sumų}}$ </span> $\Leftrightarrow$ <span style="color:red"> $\boxed{\frac{a}{b} + \frac{c}{d} \neq \frac{a+c}{b+d}}$ </span>, bet <span style="color:green"> $\boxed{\frac{a}{b} + \frac{c}{d} = \frac{ad+bc}{bd}}$ </span>
* <span style="color:red"> $\boxed{\text{Skirtumas, sudarytas iš trupmenų}\neq\text{trupmena, sudaryta iš skirtumų}}$ </span> $\Leftrightarrow$ <span style="color:red"> $\boxed{\frac{a}{b} - \frac{c}{d} \neq \frac{a-c}{b-d}}$ </span>, bet <span style="color:green"> $\boxed{\frac{a}{b} - \frac{c}{d} = \frac{ad-bc}{bd}}$ </span>
* <span style="color:red"> $\boxed{\text{Sumos kvadratas} \neq \text{kvadratų suma}}$ </span> $\Leftrightarrow$ <span style="color:red"> $\boxed{(a+b)^2\neq a^2+b^2}$ </span>, bet <span style="color:green"> $\boxed{(a+b)^2=a^2+2ab+b^2}$ </span>
* <span style="color:red"> $\boxed{\text{Skirtumo kvadratas} \neq \text{kvadratų skirtumas}}$ </span> $\Leftrightarrow$ <span style="color:red"> $\boxed{(a-b)^2\neq a^2-b^2}$ </span>, bet <span style="color:green"> $\boxed{(a-b)^2=a^2-2ab+b^2}$ </span>
* <span style="color:red"> $\boxed{\text{Šaknis iš sumos} \neq \text{šaknų suma}}$ </span> $\Leftrightarrow$ <span style="color:red"> $\boxed{\sqrt{a+b}\neq\sqrt{a}+\sqrt{b}}$ </span> (tolimesni pertvarkymai negalimi)
* <span style="color:red"> $\boxed{\text{Šaknis iš skirtumo} \neq \text{šaknų skirtumas}}$ </span> $\Leftrightarrow$ <span style="color:red"> $\boxed{\sqrt{a-b}\neq\sqrt{a}-\sqrt{b}}$ </span> (tolimesni pertvarkymai negalimi)

### Užmaršumas vienodu būdu keičiant narių grupes

Anksčiau aptarinėjome įdomų dėsnį: *padauginti sudėtus reiškia sudėti padaugintus*. Šis dėsnis užrašomas: $x(y+z) = xy + xz$. Jei atliktume vieną ar keletą įstatymų ($x \to -x$, $x \to -1$, $y \to -y$, $z\to -z$) tai gautume keletą kitų šio dėsnio variacijų, kuriose mokiniai taip pat dažnai daro klaidas.

Deja, dažna klaida būna užmiršimas, kad <span style="color:green">$\boxed{\text{pakeisti reikia visus sudedamus ar atimamus narius, o ne tik dalį jų}}$ </span>.

1\. <span style="color:red"> $4(x^2 - 10) \neq 4x^2 - 10$ </span>, <span style="color:green"> $4(x^2 - 10) = 4x^2 - 40$ </span>.

2\. <span style="color:red"> $\frac{3x^2 - x}{x} \neq 3x - x$ </span> ir <span style="color:red"> $\frac{3x^2 - x}{x} \neq 3x^2 - 1$ </span>, <span style="color:green"> $\frac{3x^2 - x}{x} = 3x - 1$ </span>. Tai buvo taisyklės variacija, naudojamas dauginimas iš $\frac{1}{x}$, kitaip tariant dalyba iš $x$. Be to $x$ atėmimas atitinka $-x$ pridėjimą.

3\. <span style="color:red"> $\frac{8 - 3}{2} \neq 4 - 3$ </span>, <span style="color:green"> $\frac{8 - 3}{2} = 4 - \frac{3}{2}$ </span>. Analogiškai ankstesniam atvejui turėjome padalinti visus narius.

4\. <span style="color:red"> $x^2+3x-5-(4x-5) \neq x^2+3x-5-4x-5$ </span>, <span style="color:green"> $x^2+3x-5-(4x-5) = x^2+3x-5-4x+5$ </span>. Čia buvo atliekama daugyba iš -1, tik pertvarkymuose tai sunkiau pastebėti. Atimti $4x-5$ reiškia pridėti $-(4x-5)$: 

<span style="color:blue"> $-(4x-5) = -1\times(4x-5) = 4x \times (-1) -5 \times (-1) = -4x + 5$ </span>

Jei paskutinės dalies aritmetinis aiškinimas atrodo painus, galime pagalvoti apie tokią situaciją: kaip pasikeistų mano santaupos $...$ po apsipirkimo, jei savo pirkinių išlaidas, lygias $4x$, sumažinčiau 5 eurais? Atsakymas: po apsipirkimo turėčiau 5 eurais daugiau ir tada santaupos būtų $...-4x+5$.

### Pražūtingi kvadratai

Šitaip kvadratu kelti negalima: <span style="color:red">$(4x)^2 \neq 4x^2$ </span>. 

Reikia kelti šitaip: <span style="color:green "> $(4x)^2 = 16x^2$ </span>

Taip yra todėl, kad pakelti kvadratu reiškia padauginti skaičių iš savęs: <span style="color:blue">$(4x)^2 = (4x) \times (4x) = 4 \times x \times 4 \times x = 16x^2$ </span>.

--------------------------------------------------------------------------------
Kitas blogo veiksmo pavyzdys:

<span style="color:red">$(-3)^2 \neq -9$ </span>. 

Reikėjo daryti šitaip: <span style="color:green "> $(-3)^2 = 9$ </span>

Aiškinimas panašus: <span style="color:blue">$(-3)^2 = (-3) \times (-3) = -1 \times 3 \times -1 \times 3 = 9$ </span>.
