### Santrauka

Darbe analizuojami tekstiniai uždaviniai, kuriems išspręsti pakanka atlikti vieną ar keletą aritmetinių operacijų: atimties, sudėties, dalybos ir daugybos, sutinkami 5 - 8 klasės mokymo programoje. Aritmetiniais veiksmais laikomos 4 aritmetinės operacijos. I darbo dalyje apibūdinamas elementaraus tekstinio uždavinio konstruktas, apimantis uždavinius, kurių sprendimuose aritmetiniai veiksmai ženklina operacijas, atliekamas su realaus pasaulio objektais. II dalyje apibūdinami subkonstruktai, pagal kuriuos šių uždavinių sprendimuose naudojamus aritmetinius veiksmus galima suklasifikuoti į atskiras kategorijas ir subkategorijas. III dalyje aprašoma ir realizuojama duomenų bazė, remiantis užklausomis siūlomos įžvalgos apie mokyklinių vadovėlių turinį. IV dalyje gauti rezultatai lyginami su kitų šaltinių rezultatais, skelbiamos viso darbo išvados.

### Įvadas <a class="anchor" id="first-bullet"></a>

Vienas didžiausių iššūkių, tenkančių moksleiviams mokykloje, yra išmokti interpretuoti tekstą. Žmogus, neturintis teksto interpretavimo įgūdžių, ne tik negeba darniai reikšti minčių ir susiorientuoti įvariose situacijose, bet negeba ir bendrauti kultūringai, tinkamai reaguodamas į pašnekovo mintis, nėra laisvas kalbos, itin svarbios žmogaus savybės, „vartotojas“, yra savo ribotumo įkaitas[1]. Matematinio lavinimo kontekste teksto interpretavimas lydi tiek mokyklinių tekstinių uždavinių sprendimą, tiek studijuojant aukštajai matematikai būdingus teoremų įrodymus, tiek remiantis kitų autorių mintimis moksliniuose tyrimuose.  

Tekstinių uždavinių sprendimas yra viena pirmųjų patirčių, su kuriomis susiduria jaunas žmogus siekamas mokyklinio, o vėliau ir akademinio išsilavinimo. Tačiau valstybinių matematikos brandos egzaminų rezultatai rodo, kad tekstinius uždavinius, kuriems išspręsti reikia bent kelių aritmetinių veiksmų, įveikia tik nedidelė dalis moksleivių. Vadinasi, daugumos moksleivių pažintis su teksto interpretavimu matematikos pamokose greičiausiai buvo nesėkminga. Kokios priežastys galėjo lemti šią nesėkmę? Galbūt egzamino uždaviniai neatitinka to, ką moksleiviai mokėsi mokykloje? O gal moksleiviai buvo nepakankamai ruošiami? 

Atsakymus bandoma atrasti nagrinėjant 5 - 8 klasių uždavinių turinį. Pirmoje darbo dalyje iš pradžių suformuluosime konstruktą, pagal kurį galima nuspręsti, kurie uždaviniai iš *Atrask* vadovėlio įeis į tyrimo apimtį, pristatysime su kokiais iššūkiais susiduria moksleiviai, spręsdami tokio tipo uždavinius. Vėliau pateiksime subkonstruktus, padėsiančius suskirstyti uždavinius į atskiras klases priklausomai nuo į sprendimą įeinančių aritmetinių veiksmų interpretacijos. 

Antroje dalyje pristatysime duomenų bazę, sudarytą iš atrinktų uždavinių ir jų savybių. Remiantis šios duomenų bazės užklausų rezultatais pateiksime įžvalgas apie dabartinį mokyklinės matematikos turinį ir pasiūlymus, ką šiame turinyje vertėtų keisti. Taip pat palyginsime uždavinius su kituose šaltiniuose aptinkamais uždaviniais: Valstybinio brandos egzamino uždaviniais, Kengūros konkurso uždaviniais, bei uždaviniais, skelbiamais lyginamuosiuose JAV ir Kinijos mokytojų žinių tyrimuose pagal Liping Ma. 

### Elementaraus tekstinio uždavinio konstruktas <a class="anchor" id="second-bullet"></a>

Penktos - aštuntos klasių Atrask vadovėliuose kiekvienoje klasėje yra pateikiama apie 10 skyrių su naujomis temomis, o kiekvieną skyrių sudaro po keletą poskyrių. Bet kurį poskyrį sudaro du puslapiai: kairiajame puslapyje pateikiama probleminė situacija, ja remiantis išaiškinama pagrindinė temos teorija, pateikiama sprendimo pavyzdžių; dešiniajame puslapyje - trijų lygių uždaviniai[2]. Šio tyrimo dėmesio centre bus tik dešiniajame puslapyje esantys neišspręsti uždaviniai. Dalis uždavinių yra palyginus nesudėtingi, pavyzdžiui, prašoma apskaičiuoti vieną dėmenį, kai duotas kitas dėmuo ir dviejų dėmenų suma (1)(5kl., 29psl.), nurodyti mažiausią galimą žiūrovų skaičių, kai jų yra apie 13000 tūkstančių tikslumu(2) (5kl., 17psl.) arba išspręsti galvosūkį su magiškaisiais kvadratais(3) (5kl., 19psl.). Tokius uždavinius intuityviai laikysime per daug paprastais palyginus su uždaviniais, kurių sprendimuose reiktų atlikti pagrindines aritmetines operacijas, ženklinančius realaus pasaulio kontekste pasitaikančių dydžių sąryšius. 

Sieksime apibūdinti tokį elementaraus tekstinio uždavinio konstruktą, kuris leistų remiantis pateiktu apibūdinimu vienareikšmiškai patvirtinti, ar uždaviniai iš Atrask vadovėlio, tame tarpe ir uždaviniai (1), (2), (3), įeina į apibrėžimo apimtį. Čia susiduriame su dar viena problema: ne visos dydžio, elementaraus sąryšio ir realaus pasaulio konteksto sąvokos turi visuotinai priimtus apibrėžimus, todėl siekiant paprastumo joms pateiksime atskirus apibūdinimus.

***Apibūdinimas 1a***. Dydis - tai palyginama objekto ar reiškinio savybė, kuriai priskiriama skaitinė reikšmė pasirenkant matavimo vienetą.

***Apibrėžimas 1b***. Sąryšis $R$ tarp dviejų aibių $X$ ir $Y$ yra Dekarto sandaugos $X\times Y$ poaibis. Jei du nariai $x\in X$ ir $y \in Y$ priklauso sąryšiui, tai sakoma, kad $x$ ir $y$ yra susieti sąryšiu $R$.

***Apibrėžimas 1c***. Ryšys tarp baigtinio skaičiaus aibių $X_1, X_2, \dots X_n$ yra Dekarto sandaugos $\displaystyle \prod_{i=1}^n X_i$ poaibis. 

Matematinėje kalboje sąryšis dar vadinamas binariuoju sąryšiu, o ryšys - $n$ - nariuoju sąryšiu, tačiau tekstinių uždavinių kontekste apsiribosime paprastesniais pavadinimais norėdami pabrėžti, kad sprendžiant tekstinį uždavinį stengiamasi suprasti, kaip jame paminėti dydžiai siejasi. Tekstiniuose uždaviniuose gali pasitaikyti tam tikras ryšys, apibrėžtas nebūtinai vienu sąryšiu. Tačiau į elementaraus tekstinio uždavinio konstruktą įtrauksime tik elementarius (nesudėtingus) sąryšius.

***Apibūdinimas 1d***. Elementarus sąryšis - tai toks sąryšis, kurio kiekvienas elementas $(x, y)$ išlaiko pastovią kažkurio vieno iš reiškinių $x-y$, $x+y$, $xy$ arba $x:y$ reikšmę. 

Priklausomai nuo reiškinio vadinsime, kad abu narius $x$ ir $y$ sieja sudėtis, atimtis, daugyba arba dalyba. 

***Apibūdinimas 1e***. Realaus pasaulio kontekstas - tai uždavinyje pateikta situacija, kurią nagrinėjant:
* uždavinio sprendime pagal bendrą susitarimą priimta ryšį tarp kintamųjų išreikšti tam tikromis matematinėmis operacijomis;
* uždavinio kintamieji yra tam tikri pavadinimus turintys dydžiai.

Kontekstas apibrėžia, apie kokius objektus kalbama uždavinyje. Dydžiai - realaus pasaulio objektai. Tačiau skaičius nelaikytinas realaus pasaulio objektu, nes neturi konkretaus matavimo vieneto ir todėl neįeina į dydžio apibrėžimą. 

***Apibūdinimas 2*** Realaus pasaulio konteksto uždavinys, kuriame dydžius sieja ryšys, išreiškiamas elementariais sąryšiais, laikomas elementariu tekstiniu uždaviniu (ETU).

Remiantis šiuo apibūdinimu pavyzdžiai (1) ir (3) netenkina ETU apibrėžimo, nes uždavinyje pateikti kintamieji nėra realaus pasaulio objektai. Tuo tarpu pavyzdyje (2) yra realaus pasaulio kontekstas, tačiau ETU apibrėžimas netenkinamas, nes prašoma rasti mažiausią galimą tam tikro dydžio skaitinę reikšmę, tenkinančią sąlygą, kad ją suapvalinus tūkstančių tikslumu, gausime 13000. Nei apvalinimas, nei mažiausios reikšmės radimas nėra išreiškiami elementariais sąryšiais.





### Dydžio apibrėžimo korektiškumo klausimas <a class="anchor" id="third-bullet"></a>

Dydžių įvairovė plati. Fizikinių dydžių, tokių kaip atstumas, plotas, tūris, greitis, masė, laiko intervalas, temperatūra matavimo vienetai įprastai pateikiami pagal visuotiniu susitarimu priimtą etaloną. Kitais atvejais dydis dažnai būna susijęs su objektų: picų, obuolių, žmonių, geometrinių figūrų ir pan. skaičiavimu, todėl matavimo vienetais laikomi tam tikri objektai.

Jei atliekamas veiksmas yra sudėtis ir atimtis, tai tiek abu nariai, tiek rezultatas ženklina skaitines išraiškos dydžių, reiškiamų tais pačiais matavimo vienetais. Daugybos ir dalybos atveju situacija sudėtingesnė. Lygybės $b\times c = a$ ir $c = \frac{a}{b}$, $b\neq 0$ yra ekvivalenčios. Tačiau kintamasis $a$ ne visais atvejais yra dydis. Tai priklauso nuo trupmenos interpretavimo[9]. A. Asghari siūlo tokius apibrėžimus:

* **Vieneto ekvivalentiškumas** yra trupmenų lygumas, kai jos atitinka tą pačią vienodos visumos dalį. 
* **Proporcijos ekvivalentiškumas** yra trupmenų lygumas, kai santykis tarp skaitiklio ir vardiklio yra vienodas (rezultatas laikomas neturinčiu vienetų)

***Pavyzdys 3a***. Tarkime vienas žmogus suvalgė vieną iš trijų bananų, o kitas - dvi iš šešių picos skiltelių. Koks suvalgyto maisto dydis?

***Pavyzdys 3b***. Tarkime vienas žmogus suvalgė vieną iš trijų turimų pyragaičių, o kitas - du iš šešių picos skiltelių. Koks suvalgyto maisto dydis?

Pavyzdyje 3a abu žmonės suvalgė tą pačią trupmeninę dalį savo maisto. Ši dalis nėra matuojama tais pačiais vienetais, todėl trupmeninės dalies dydis gali būti skirtingas, nepaisant to, kad lygybė $\frac{1}{3} = \frac{2}{6}$ galioja. Lygybėje vienetai nerašomi, nes skaičiuojamos ne tos pačios visumos dalys. Šiuo atveju kalbama apie proporcijos ekvivalentiškumą.

Pavyzdyje 3b vienetai yra vienodi: išlaikoma ne tik skaitinių reikšmių lygybė, bet ir dydžių lygumas. Lygybėje matavimo vienetai rašomi: $\frac{1}{3}\text{ pyragaičių} = \frac{2}{6} \text{ pyragaičių}$. Šiuo atveju kalbama apie vieneto ekvivalentiškumą.

Matome, kad, jei kiekvieno nario $(a,b)$, priklausančio elementariam sąryšiui $X$, kintamuosius $a$ ir $b$ sieja dalyba ($a:b$ reikšmė yra pastovi), tai rezultatas $a:b=c$ netenkina dydžio apibūdinimo 1a ir nėra realaus pasaulio objektas. Dėl šios priežasties ekvivalenčioje lygybėje $b\times c = a$ kintamasis taip pat nėra realaus pasaulio objektai - apibūdinimas 1e tampa nekorektišku. Norint išspręsti šią problemą daugelyje mokslo sričių naudojama bedimensinio dydžio sąvoka. Tokiems dydžiams yra deklaruojamas matavimo vienetas 1, įprastai neįeinantis į dydžio žymėjimą.

### Elementaraus tekstinio uždavinio sprendimo proceso sudedamosios dalys <a class="anchor" id="fourth-bullet"></a>

Natūralu, kad moksleivio pastangos didina galimybes pasiekti aukštesnius mokymosi rezultatus pademonstruojant įvairias mokykloje įgytas kompetencijas, tame tarpe ir tekstinių uždavinių sprendimo gebėjimus. Kai kurie autoriai pažymi, kad mokymosi metu susiduriant su matematiniu turiniu mokymosi procesas nėra vientisas. George Pólya savo knygoje How To Solve It tvirtina, kad besimokantysis turėtų įgyti tiek patirties, kiek tai gali padaryti savarankiškai. Jei jis yra paliekamas be pagalbos arba su per maža pagalba, pažangos nepadarys. Jeigu jam būtų padedama per daug, indėlis į pažangą būtų per mažas. Vadinasi, mokytojai turėtų moksleiviams padėti padėti nei per mažai, nei per daug tam, kad besimokančiojo ir mokintojo darbas pasiskirstytų tam tikrose ribose. 

Mokymosi procese svarbu ne tik mokytojo ir moksleivio bendradarbiavimas, bet ir mokomosios medžiagos turinys. Mokytojai dažnai nėra pajėgūs identifikuoti konkrečių kiekvieno moksleivio trūkumų, stabdančių pažangą ir įvertinti, kas turėtų įeiti į mokomąjį turinį padedantį šiuos trūkumus pašalinti. Šioje dalyje pailiustruosime, kaip atrodo ETU sprendimo procesas ir kas gali daryti įtaką ETU sudėtingumui.

**Pavyzdys 1a**. Man dabar dvigubai daugiau metų nei Jums buvo tada, kai man buvo tiek metų, kiek Jums dabar. Mums abiems kartu 35 metai. Kokie mūsų amžiai?

Pirmas sprendimo būdas galėtų būti mano ir Jūsų amžių pažymėti nežinomaisiais $x$ ir $y$. Antroji sąlyga nesudėtinga: žodis ,,kartu" siūlo užuominą, kad sudėję abu nežinomuosius gausime kitą dydį, kuriam priskiriamas skaičius $35$. Todėl gauname, kad nežinomieji $x$ ir $y$ tenkina elementarųjį sąryšį, kurį ženklina lygybė $x+y=35$. Likusioje sąlygos dalyje siejami laiko atžvilgiu kintantys dydžiai, o sprendimą apsunkina nestandartinis atvejis, kuomet dydžiams priskiriamos reikšmės ne tuo pačiu laiko momentu. Norint suprasti, kaip sąryšį pateikti aiškiau, galima nagrinėti tokį pagalbinį uždavinį:

**Pavyzdys 1b**. Man dabar 25 metai, o Jums - 20. Kiek man metų buvo tada, kai man buvo tiek metų, kiek Jums dabar? 

Dabar matyti, kad kalbama apie laikotarpį prieš $25 - 20 = 5$ metus. Grįžkime prie ankstesnio pavyzdžio. Naudojantis anksčiau įvestais nežinomaisiais, galima užrašyti, kad minėtas laikotarpis buvo prieš mūsų amžių skirtumą: $x-y$ metų. Jei dabar Jums $y$ metų, tai tuo laikotarpiu jums buvo $y-(x-y)$ metų. Liko neišnaudota tik gana paprasta sąlygos dalis, kurioje yra žodis ,,dvigubai". Šis žodis siūlo užuominą, kaip užrašyti sąryšį tarp dviejų dydžių: $x = 2\left(y-(x-y)\right)$

**Pavyzdys 1c**. Išspręskite lygčių sistemą: $\begin{cases} x = 2\left(y-(x-y)\right) \\ x+y=35 \end{cases}$

Išsprendę šią lygčių sistemą randame nežinomųjų $x$ ir $y$ reikšmes: $20$ ir $15$. Šio pavyzdžio apimtyje nėra pasakyta, kokie ženklinamų dydžių pavadinimai, vadinasi suformuluotame uždavinyje nebuvo pateiktas realaus pasaulio kontekstas. Tačiau lygčių sistemos sprendimas galimai įeina į Pavyzdžio 1 sprendimą. 

Atkreipkime dėmesį, kad ETU lygčių arba lygčių sistemų sprendimo ir sąryšius nusakančių formuluočių keitimo procesai yra panašūs. Galima išskirti tokias bendras savybes:

1\. Invariantiškumas. Tiek sąryšiai, tiek lygties ar lygčių sistemos sprendinių aibės yra nekintamos.

2\. Redukuojamumas. Sąlygos formuluotę stengiamasi pakeisti taip, kad būtų galima lengviau atpažinti aritmetines operacijas, siejančias dydžius, o lygtis taip, kad, kad tolimesnis sprendinio gavimas supaprastėtų.

3\. Baigtinumas. Po baigtinio kiekio pakeitimų procesas turi užsibaigti aiškia išvestimi. Lygčių atveju įprasta pateikti lygybę, nurodančią lygties sprendinį. Keičiant sąryšius nusakančias formuluotes įprasta arba išsiaiškinti, koks veiksmas sieja duotus dydžius arba užrašyti lygtį, įprasminančią jų ryšį.

Dar daugiau, šie panašumai pakankamai dažnai lemia ir pasirinkimo tarp kelių skirtingų ETU sprendimo būdų galimybę.

Pavyzdys 2. Joniukas ir Grytutė turėjo po lygiai riekelių. Kai Joniukas Grytutei atidavė 5 duonos riekeles, jis turėjo dvigubai mažiau riekelių už Grytutę. Kiek riekelių kiekvienas turėjo iš pradžių?

Pažymėjus nežinomuoju pradinį Grytutės riekelių skaičių $x$ duotus kintamuosius (atiduotų riekelių bei galutiniai Grytutės ir Joniuko riekelių kiekį) sieja ryšys, užrašomas lygtimi $2(x-5)=x+5$. Sprendžiant šią lygtį būtų gaunamas sprendinys $x=15$. Lygties sprendimą galima pakeisti samprotavimu. Joniuko riekelių skaičius sumažėjo tiek, kiek Grytutės padidėjo, vadinasi susidarė skirtumas, lygus dvigubam atiduotų riekelių skaičiui, t.y. 10. Tai yra antra tiek, kiek galiausiai liko pas Joniuką. Vadinasi, pas Joniuką liko 10 riekelių, o jiedu prieš tai turėjo po 15 riekelių.

Pavyzdžio 1a sudėtingiausia sprendimo dalis buvo pastebėti, kaip ryšį tarp dabartinio laikotarpio ir laikotarpio, kai man buvo tiek metų, kiek Jums dabar, išreikšti elementariais sąryšiais. Norint įgyti kompetenciją, kaip spręsti tokius ETU, galime rinktis šias išeitis:

Išeitis 1. Studijuoti ETU su analogiškomis sąlygomis.

Išeitis 2. Studijuoti ETU, kuriuose pagalbinio uždavinio suformulavimas ir išsprendimas padeda įžvelgti, kaip ryšys tarp įvairių dydžių reiškiamas elementariais sąryšiais

Pagal A. Romašką matematikų tikslas turėtų būti maksimizuoti rezultatų bei laiko sąnaudų naudingumą, o tai padėtų padaryti objektyvesnis požiūris į mokslo ribas bei sugebėjimas taikyti paprastus modelius problemoms spręsti[4]. Išeitis 1 siūlo kiekvienai nenumatytai problemai kurti atskirą modelį, o tai būtų ir beribis, ir rizikingas procesas turint mintyje ribotą mokyklinėje programoje numatytą mokymosi laiką. Tuo tarpu Išeitis 2 būtų tikslingenė: ji leistų pamatyti, kas sieja skirtingas problemines situacijas, būtų bandoma ieškoti kelių, supaprastinančių sudėtingų uždavinių sprendimą, kai nėra iš anksto žinoma tinkančių teorinių modelių.

Išeitis 2 taip pat lavina matematinę intuiciją - galėjimą įgyti žinias nesiremiant sąmoningu samprotavimu[5]. G. Pólya pateikia sąrašą veiksmų, netiesiogiai susijusių su mintinėmis operacijomis, įprastai pasitarnaujančiomis uždavinių sprendimui. Šie veiksmai gali būti išdalinti į 4 pagrindines kategorijas: susipažinimą su uždaviniu, sprendimo plano sudarymą, sprendimo plano realizavimą bei sprendimo apžvalgą. Sudėtingiausios pavyzdžio 1a sprendimo dalies fragmentas atitinka dalį G. Pólya pateikto plano sudarymo:

*Pabandyk sugalvoti pažįstamą uždavinį, turintį tą patį ar panašų nežinomąjį. Tarkim radai panašų prieš tai išspręstą uždavinį. Ar gali juo pasinaudoti? Ar gali panaudoti jo rezultatą, sprendimo metodą?*

Nors sprendžiant ETU svarbus tiek lygčių ir lygčių sistemų sprendimo gebėjimas, tiek matematinė intuicija, šių dviejų ypatybių toliau netyrinėsime.

Nežinomuosius dydžius siejančių aritmetinių veiksmų paieškos gali būti sudėtingesnės už vien aprašytąjį intuityvų mąstymą. Antai uždavinyje, kur sudedant valties savajį greitį su upės greičiu gaunamas valties greitis plaukiant pasroviui, galioja pagal bendrą susitarimą priimtas ryšys tarp trijų dydžių. ### Trupmenų subkonstruktai <a class="anchor" id="sixth-bullet"></a>
[Plačiau](https://link.springer.com/article/10.1007/s10649-021-10030-7)

Tarp praeito amžiaus XIII - ojo ir IX - ojo dešimtmečių Kieren (1976), Vergnaud (1983) and Freudenthal (1983) nepriklausomai plėtojo įvairius trupmenų subkonstruktus, aspektus ir objektus. Pradiniame tyrimų etape Kieren išskyrė 4 trupmenų subkonstruktus: truomeną kaip dalmenį, kaip matavimą, kaip operaciją ir kaip santykį. Vėlesni jo darbai pasipildė dar vienu, trupmeno kaip dalies - visumos subkonstruktu, kuriame truomena interpretuojama kaip viena ar kelios lygios vieneto arba visumos dalys. 1983 m. Rationaliųjų Skaičių Projekte buvo pasiūlyta šį subkonstruktą laikyti persidengiančiu su visais likusiais subkonstruktais. Čia išvardysime likusių subkonstruktų apibūdinimus:

* **Quotient (dalmuo)**. Trupmena interpretuojama kaip dalybos rezultatas: 3:4 = 3/4. Skaitiklis gali būti visumos dydis, o vardiklis gavėjų kiekis (partitative division context). Taip pat skaitiklis gali būti visumos dydis, o vardiklis vienos visumos dalies dydis (measurement division context).
* **Measurement (matavimas)**. Šiame subkonstrukte trupmena yra glaustai susijusi su dviem savokomis: skaičius, atitinkantis tašką skaičių ašyje, ir skaičius, atitinkantis atkarpos ilgį. Pagal pirmą interpretaciją vardiklis nurodo, į kiek dalių yra dalijama vienetinė atkarpa, o skaitiklis - šių dalių skaičių. Pagal antrą interpretaciją vardiklis nurodo, į kiek dalių yra dalijamas tam tikras intervalas, o skaitiklis - to intervalo ilgį. 
* **Operation (operacija)**. Trupmena yra susijusi su procesu, vykstančiu keičiant objektą į kito dydžio objektą. Operatorius suvokiamas kaip funkcija, taikoma tam tikram objektui. Skirstomas į stretcher/shrinker operator ir duplicator/partition-reducer operator. Stretcher/shrinker operator nekeičia vienetų skaičiaus, o bet keičia jų didumą. Duplicator/partition-reducer operator atvirkščiai.
* **Ratio (santykis)**. Trupmena išreiškia dviejų dydžių palyginimą. Lyginimo rezultatas yra palyginimo rodiklis, simbolizuojantis proporciją. Proporcija taip pat aprašo ekvivalentumo sąryšį. Kiekvieniems 7 nepasiskiepiujisiems tenka 3 pasiskiepijusieji. 

Vėlesniame darbe nagrinėdami uždavinius dalies - visumos subkonstrukto neįtrauksime.

**Trupmenos: senesnė versija**

|Subkonstruktas|Sutartinis žymėjimas|Kintamieji|Pirma interpretacija|Antra interpretacija|
|:---:|:---:|:---:|:---:|:---:|
|Quotient|$a:b$|(a,b)|(visuma, visumos dalių kiekis)|(visuma, visumos dalies dydis)|
|Measurement|$a/b$|(a,b)|(dalių kiekis, dalių kiekis vienetinėje atkarpoje)|(panaudotų dalių kiekis intervale, intervalo ilgis)|
|Operation|$a*(b/c)$|(a,b,c)|(objektų kiekis, objekto dydžio padidinimas, objekto dydžio sumažinimas)|(objekto dydis, objektų kiekio padidinimas, objektų kiekio sumažinimas)|
|Ratio|$a:b$|(a,b)|(kiekis, kiekis) arba (dydis, dydis)|

**Trupmenų subkonstruktus aprašanti lentelė su Pandas biblioteka**

```python
import numpy as np
arr1 = np.array([['Quotient','Measurement','Operation','Ratio'], 
                   ['$a:b$', '$a/b$', '$a\\times(b/c)$', '$a:b$'],
                   ['(a, b)', '(a, b)', '(a, b, c)', '(a, b)']])
arr2 = arr1[:,[0,0,1,1,2,2,3]]
arr2 = np.r_[arr2, [['Partitative division','Measurement division','Interpretation 1','Interpretation 2','Stretcher/shrinker', 'Duplicator/partition-reducer','']]]
arr2 = np.r_[arr2, [['$a$ - visumos dydis; $b$ - gavėjų kiekis',
                     '$a$ - visumos dydis; $b$ - dalies dydis',
                     '$a$ - panaudotų dalių kiekis; $b$ - dalių kiekis vienetinėje atkarpoje',
                     '$a$ - dalių kiekis intervale; $b$ - intervalo ilgis',
                     '$a$ - objektų kiekis; $b$ - objekto dydžio padidinimas; $c$ - objekto dydžio sumažinimas', 
                     '$a$ - objektų dydis; $b$ - objektų kiekio padidinimas; $c$ - objektų kiekio sumažinimas',''
                     '$a$ ir $b$ - bet kokie vienodą santykį išlaikantys dydžiai']]]
df_fractions = pd.DataFrame(arr2.T, columns=('Subkonstruktas', 'Veiksmas', 'Kintamieji', 'Tipas', 'Kintamųjų paaiškinimai'))
df_fractions.set_index(['Subkonstruktas', 'Veiksmas', 'Kintamieji', 'Tipas'], inplace=True)
df_fractions
```

### Atimties kategorijos <a class="anchor" id="seventh-bullet"></a>

Kategorijos, skirtos klasifikuoti uždaviniams, kuriuose atimami tam tikri dydžiai, buvo pasiūlytos apie 1978m. Heller ir Greeno darbuose: keitimas, sujungimas ir palyginimas. Heller and Greeno (1978). Vėliau Carpenter ir Moser (1981) pasiūlė įtraukti taip pat ir sulyginimo kategoriją. Norint lengviau suprasti uždavinių, kuriuos atlikti galima su sudėtimi arba atimtimi, klasifikavimo procesą, patogu apibrėžti, ar kintamieji yra statiški, ar dinamiški.

Jei kintamieji dinamiški:

* **Change (keitimas)** atveju kalbame apie tam tikrą pradinį dydį, kuris su tam tikro veiksmo pagalba buvo transformuotas į galutinį dydį. Transformavimas gali duoti dvejopą pokytį: sumažėjimą ir padidėjimą.
* **Equalize (sulyginimas)** atveju siūloma, vieną dydį pakeisti, kad jis taptų lygus antrajam. Pirmąjį dydį vadinsime stabiliu, o antrąjį keičiamu.

Jei kintamieji statiški:

* **Combine (sujungimas)** atveju kalbama apie tam tikrą objektą (visumą), sudarytą iš kitų atskirų objektų (visumos dalių). 
* **Compare (palyginimas)** atveju vartojamas lyginimas: *vienas dydis mažesnis ar didesnis už kitą*. Pirmąjį dydį vadinsime lyginamu (originaliai: referent quantity), o antrąjį lygintinu (originaliai: compared quality). 

Taip pat naudosime tokias sutartines sąvokas:

* Dinamiškų kintamųjų atveju neigiamu pokyčiu vadinsime dydį, nurodančią, kiek sumažėjo pradinis arba keičiamas dydis, o teigiamu - kiek padidėjo. 
* Statiškų kintamųjų atveju neigiamu skirtumu vadinsime dydį, nurodantį, kiek antras dydis mažesnis už pirmą, o teigiamu - kiek didesnis. 
 
Pateiksime siūlomą visų tekstinių uždavinių, kurių sprendimas reikalauja vien tik dviejų skaičių atimties $a-b$, klasifikaciją. Pagal nutylėjimą, skaitinės $a$, $b$ ir $a-b$ reikšmės yra teigiamos.

**Skaičių atimties subkonstruktus aprašanti lentelė su Pandas biblioteka**

### Juodraštis, kuriame eksperimentuojama, kaip turėtų būti aprašytas kodas, skirtas duomenų bazės valdymui

```python
from sympy import *
import re
import string

a,b,c,d,e,f,g,h,i,j,k,l,m,n,o,p,q,r,s,t,u,v,w,x,y,z = symbols('a b c d e f g h i j k l m n o p q r s t u v w x y z')
class Evaluator:
    def __init__(self, expr):
        self.expr = expr
    
    def set_numerics(self):
        vars_numeric = re.findall(r'\d+', self.expr) #list of numeric values identified (d instead of w)
        vars_numeric_noduplicates = list(dict.fromkeys(vars_numeric)) #list of distinct numeric values
        self.numerics = vars_numeric_noduplicates
        self.is_suspicious = len(vars_numeric) != len(vars_numeric_noduplicates)
        
    def set_unknowns(self):
        self.set_numerics()
        self.unknowns =dict(zip(string.ascii_lowercase, self.numerics))
        self.unknowns_reverse = dict(zip(self.numerics, string.ascii_lowercase))
        
    def set_form(self):
        self.set_unknowns()
        pattern = re.compile("|".join(self.unknowns_reverse.keys()))
        self.form = pattern.sub(lambda m: self.unknowns_reverse[re.escape(m.group(0))], self.expr)
        
expr = '9*x-787*(564+x)+9'       
ev = Evaluator(expr)
ev.set_form()
print(f'expression = {expr}')
print(f'unknowns = {ev.unknowns}')
print(f'unknowns_reverse = {ev.unknowns_reverse}')
print(f'form = {ev.form}')
print(f'is it suspicious: {ev.is_suspicious}')
```
