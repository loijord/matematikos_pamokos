### Duomenų bazės pavyzdys
Pasibaigus pamokai į duomenų bazę buvo suvesti darytų užduočių ir jų dalių atlikimo rezultatai.

### Pamokos peržiūra
Joje galima pamatyti:
* Kokie gebėjimai buvo testuoti
* Kaip atrodė uždavinių formuluotės

### Ką nuveikėme

* Visą informaciją apie uždavinius ir jų sprendimus galima rasti [nuorodoje](susitikimas1.ipynb). Jeigu nuoroda neatsidaro, reikia eiti į [https://nbviewer.jupyter.org](https://nbviewer.jupyter.org), nukopijuoti nuorodos adresą į ten esantį laukelį ir paspausti **Go!**
* Valandą laiko skyrėme apatarti:
 * ką reikia žinoti matematikos egzaminui
 * kaip atrodo abiturientų vertinimas
 * kaip po egzamino analizuojami abiturientų rezultatai
 * kaip atrodo egzaminą sudarančių gebėjimų pasiskirstymas
 * kodėl nukrito mokinių matematikos pasiekimai per 20 metų
 * kaip vertinamos moksleivio žinios pagal Simono sukurtą sistemą
 * su kokiais sunkumais susiduriame mokymosi metu, kai norime, kad dabartinės žinios pasistūmėtų į priekį
 
Kitą laiką aiškinomės su uždavinių sprendimu susijusius dalykus
* Per pamoką bandėme atlikti mintinį eksperimentą - pavaizduoti kreivę, simetrišką funkcijos $f(x)=cos(x)$ grafikui tiesės $y=x$ atžvilgiu. Sekėsi sunkiai, nes prireikė intelektinių gebėjimų. Šių gebėjimų reikėjimas kai kuriose matematikos srityse trukdo įgyti įvairių matematikos disciplinų žinias per trumpą laiką. Kitaip sakant, nepakanka mokytis daug formulių ar taisyklių, bet dar ir prireikia jas įsiminti talpesnėmis atminties struktūromis.
* Trikampio ploto formulę Kipras apibūdino tiksliai, tačiau Pitagoro teoremos įrodyme taikė ją labai nedrąsiai.
* Kas yra Pitagoro teorema, Kipras suformuluoti negalėjo.
* Toliau susipažinome su geometrinio uždavinio sprendimo struktūra. Jos apibūdinti Kipras pats negalėjo. Kai išsprendėme pirmą užduotį, visi sprendimo žingsniai atrodė aiškūs, tačiau nuoseklus uždavinio sprendimas atrodė sudėtingas, nes Kipras bijo uždavinių su įrodymais.
* Nagrinėjome Pitagoro teoremos įrodymą - 3 žingsniai savarankiškai nepavyko, 3 pavyko.

### Kokybinė tyrimo dalis

Moksleiviui dažniausiai nesiseka iki galo išspręsti daugelio uždavinių, todėl pamokos metu jis yra klausinėjamas ir tokiu būdu paaiškėja, kiek jis turi gebėjimų atlikti atskiras dalis.

Kol kas duomenų buvo nedaug, pagrinde iš geometrijos. Stringa žingsniai, kuriuose reikia įrodymo elementų (trikampių lygumo požymiai, kampų rūšys, kai dvi lygiagrečios tiesės yra perkirstos trečiąja). Trikampių ir stačiakampių plotas, skaidymai dauginamaisiais - labai lengvos temos, kur reikia taikyti formules, tad lengvai ir sekėsi, tačiau jų nepakanka.

Įdomus pastebėjimas: trikampio formulė $S=ah/2$ - ne tik simbolių rinkinys, bet egzistuoja ir jų prasmė. Kipras tesingai pateikė mintinį eksperimentą, norint ją išvesti (stačiakampio dalinimas į 2 dalis). Didelė dalis egzamine reikalingų taisyklių taip pat gali būti gauta naudojant panašų informacijos įsiminimą talpesnėmis atminties struktūromis. 

Norint susipažinti su kitais eksperimentais, padedančiais išvesti geometrijos formules, galima pasižvalgyti Geogebra pagrindu [suprogramuotomis vizualizacijomis](https://www.geogebra.org/materials)[galima eiti į Math -> Geometry -> Plane figures]. Taip pat šios programos pagrindu yra sukurtas [įdomus žaidimas](https://www.euclidea.xyz), ugdantis geometrinį pastabumą, kurio prireiks ir egzamine.
