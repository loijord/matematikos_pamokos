### Apie sakinį TRIJŲ VIENETŲ SUMA YRA TRYS
Savo ***Facebook*** paskyroje pamėginau apibendrinti prieš tris savaites vykusio seminaro diskusiją. Antanas pasiūlė pagalvoti, kokios trys klaidos yra sakinyje TRIJŲ VIENETŲ SUMA YRA TRYS.

***Tokių klaidų prigimtis yra aiškių matematinių sąvokų apibrėžimų neturėjimas. Seminare generavome idėjas, kokie turėtų būti dydžio, kiekio ir matavimo vieneto apibrėžimai. Lyg ir gavosi, kad jie yra nepaklūstantys bendrai tvarkai, nes norint apibrėžti bent vieną iš jų, prireikia kitų dviejų ir taip susidaro uždaras ratas. Rimas patvirtino, kad tame blogybių nėra. Aš priminiau, kad dydis - tai objekto ar reiškinio savybė, kuriai įprasta priskirti skaičių (kiekį) pasirenkant matavimo vienetą. Kiekis, kiekybė - tai, kas išreiškiama skaičiais, suskaičiuojama. Tačiau dėl vieneto liko nelabai aišku. Pagal mokytoją Antaną, vienetas - tai objektų kiekis, sudarantis tam tikrą visumą, o prie jo prijungus kiekį, gaunam matavimo vienetą. Pvz. ir viena avis, ir pora medžių, ir trejetas avių būtų vienetai. O pagal mane (ir kitus) matavimo vienetas - tai būtų bendru sutarimu apibrėžtas dydis, naudojamas kaip etalonas. Bendrai sutarėme, kad dydis atsako į klausimą ,,kas matuojama?", o matavimo vienetas - ,,kuo matuojama?".***

Į šią ištrauką galima žiūrėti kaip į teorinę medžiagą, pakankamą atsakyti Antano klausimui. Jis taip pat pridūrė, kad tada kai sąvokų apibrėžimuose paliekami užkulisai, gaunasi nesusipratimai ir šią mintį minėtu klaidingu sakiniu. Remiantis pateikta medžiaga suformulavau tokius atsakymus.

1\. Vienetai negali būti trys, nes vienetas - visuma. Mes nesakome ,,trys poros kiškių", o vietoj to sakome ,,šešetas kiškių" (pora kiškių pagal Antaną - vienetas).

2\. Vienetai įprastai nesumuojami

3\. Tai, kas sumuojama ir kas po to gaunama, turi būti tos pačios kategorijos objektai. Vienetas ir trys - skirtingų kategorijų objektai.

**Klausimas.** Kurioje schemos dalyje turėtų būti patalpinta Antano pasiūlyta samprata?

### Kiti klausimai seminaro dalyviams

1\. Ar tinka sakyti, kad trys - tai vieneto žymėjimas? Reikia pasakyti kokio vieneto.

2\. Lygybės sąvokos aiškinimas kiek modifikuotas lyginant pagal originalų tekstą. Nesupratau, ką reiškia teiginys $2+3$ ir $1+4$ yra du skirtingi to paties skaičiaus vardai (reprezentacijos). Skaičius kaip objekto tipas.

3\. Sudedamosios dalys sąvokai apibūdinti: apibrėžimas, žymėjimas, vaizdavimas.

4\. Tarkime paprašėme moksleivio apibūdinti trupmenos prasmę $\frac{1}{3}$ arba paaiškinti, kodėl galime tvirtinti, kad ji lygi $\frac{1\cdot 5}{3 \cdot 5}$. Moksleivis nurodo, kad prasmė būtų, jog 1 - tai trupmenos skaitiklis, o 3 - vardiklis. Į antrą klausimą jokio atsakymo neturi išskyrus, kad taip mokėsi mokykloje. Kaip elgtumėtės tokioje situacijoje? Reikėtų klausti ne kokia prasmė, o prašyti nurodyti $\frac{1}{3}$ apibrėžimą.
